```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

colorcycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

### Exemplo 3.12

```python
xi = np.array([0.450, 0.467, 0.431, 0.440, 0.452, 0.458, 0.438, 0.462, 0.447, 0.452])

X_bar = np.mean(xi)
sx2 = np.var(xi, ddof = 1)
sx = np.sqrt(sx2)

n = len(xi)
```

```python
muX_inf = scipy.stats.t.ppf(0.025, n-1, loc = X_bar, scale = sx/np.sqrt(n))
muX_sup = scipy.stats.t.ppf(0.975, n-1, loc = X_bar, scale = sx/np.sqrt(n))

print(muX_inf,'< Mu_x <', muX_sup)

fig, ax = plt.subplots()

x = np.linspace(0.43,0.47, 100)
ax.plot(x, scipy.stats.t.pdf(x, n-1, loc = X_bar, scale = sx/np.sqrt(n)))

ax.axvline(x = muX_inf, ls='--', color = colorcycle[1])
ax.axvline(x = muX_sup, ls='--', color = colorcycle[1])
ax.axvline(x = X_bar, ls='-', color = 'k')
ax.set_title('Intervalo de 95% de confiança para a média --  t de student');
```

### Exemplo 3.13

```python
muX_inf = scipy.stats.norm.ppf(0.025, loc = X_bar, scale = sx/np.sqrt(n))
muX_sup = scipy.stats.norm.ppf(0.975, loc = X_bar, scale = sx/np.sqrt(n))

print(muX_inf,'< Mu_x <', muX_sup)

fig, ax = plt.subplots()

x = np.linspace(0.43,0.47, 100)
ax.plot(x, scipy.stats.norm.pdf(x, loc = X_bar, scale = sx/np.sqrt(n)))

ax.axvline(x = muX_inf, ls='--', color = colorcycle[1])
ax.axvline(x = muX_sup, ls='--', color = colorcycle[1])
ax.axvline(x = X_bar, ls='-', color = 'k')
ax.set_title('Intervalo de 95% de confiança para a média -- normal');
```

### Exemplo 3.15
