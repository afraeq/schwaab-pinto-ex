```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats

colorcycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

### Exemplo 4.12

```python
y_e = np.array([1050, 2000, 2950, 4000])
x_e = np.array([10, 20, 30, 40])
```

```python
def linear_regression_4_12 (x, y):
    alpha = np.sum(x*y)/np.sum(x**2)
    return alpha*x, alpha

print('y_e: ',y_e)
y_m, alpha = linear_regression_4_12(x_e, y_e)
print('y_m: ', y_m)
e = y_e-y_m
print('e  : ', e)
```

```python
sigmay2 = 3000

sigmaE2 = np.sum(e**2)/(len(e)-1)
sigmaE2
```

```python
F_inf = scipy.stats.f.ppf(0.025,3,1e10)
F_sup = scipy.stats.f.ppf(0.975,3,1e10)

print('F = ',sigmaE2/sigmay2)
print(F_inf,'< F <', F_sup)
```

```python
fig, ax = plt.subplots(figsize = (15,8))

x = np.linspace(0,50,100)
y = alpha*x
ax.plot(x,y, color='k')

ax.errorbar(x_e, y_e, yerr = 2*np.sqrt(sigmay2), capsize = 10, ls='', marker = '*', color='k')
#plt.plot(x_e, y_m, '-')
ax.axis([0,50,0,5000])
ax.grid('on')

ax.plot(x,y+2*x,'--', color='k')
ax.plot(x,y-2*x,'--', color='k');
```
